## Especificación del Modelo. 
El modelo aumentado de Solow de Mankiw-Romer-Weil (MRW) se utiliza como base para este trabajo. La función de producción que incorpora el tamaño y la calidad del gobierno es de la forma Cobb-Douglas: 

(13) $y(t)=Ak(t)^α  h(t)^β  (g(t) e^(µθ) )^γ $

donde y(t) es la producción por trabajador; k(t) es el stock de capital privado por trabajador; y h(t) es el stock de capital humano por trabajador. La dimensión de tamaño del gobierno g(t), se mide por el stock de capital gubernamental por trabajador, y θ mide la dimensión de calidad del gobierno. Se asume la forma exponencial para la variable de calidad, ya que la buena gobernanza no es un factor de entrada directo, sino que sirve para mejorar la eficiencia del stock de capital del gobierno.

En el equilibrio de estado estacionario, hay una tasa exógena de progreso tecnológico ϖ y una tasa de crecimiento de la fuerza laboral n. El stock de capital se deprecia a una tasa δ. Si la fracción del ingreso dedicada al capital privado, el capital humano y el capital público son respectivamente, SK, SH y SG, el nivel de estado estable de la producción per cápita se alcanza cuando la adición al stock de cada tipo de capital es solo suficiente para satisfacer la tasa de crecimiento de la fuerza laboral y reemplazar el capital que se deprecia a una tasa de δ. El nivel de estado estacionario de producción per cápita en forma lineal logarítmica se puede expresar:
(14) $ln⁡[Y(t)/L(t) ]=a_0+a_(1 ) ln[S_K/(n+ϖ+ δ)]+a_2  ln[S_H/(n+ϖ+ δ)]+a_3  ln[S_G/(n+ϖ+ δ)]+a_4  θ $
Donde a_4=µγ.

La tasa de crecimiento del producto por trabajador en la transición al estado estacionario se puede expresar: 
(15) $ln⁡y(t)-ln⁡ y(0)=(1-e^(-λt))[ln⁡(y^* )-ln⁡y(0) ]$
donde y(0) es el nivel inicial de producción por trabajador e y* es el nivel de ingresos en estado estacionario por trabajador. λ es la velocidad de convergencia y además λ=(1-α-β-γ)(n+ϖ+ δ). Restando y(0) de ambos lados de la ecuación y sustituyendo por y* se obtiene el modelo de transición que se puede estimar como:

(16)  $ln⁡ y(t)-ln y(0)= a_0+a_(1 ) ln[S_K/(n+ϖ+ δ)]+a_2  ln[S_H/(n+ϖ+ δ)]+a_3  ln[S_G/(n+ϖ+ δ)]+a_4  θ+a_(5 )  ln⁡ y(0)+ ε  $

De acuerdo con la ecuación (16), la tasa de crecimiento del ingreso per cápita depende de la acumulación de capital privado, capital humano, capital público y de un buen gobierno. Aplicando el mismo razonamiento que Hulten (1996), un país con mejor gobierno convergerá a un mayor nivel de ingreso per capita estable que un país con mala calidad gubernamental. Si comienzan con el mismo nivel de ingreso per cápita, el país con una mejor gobernanza experimentará una tasa de crecimiento más rápida. 

## Metodología

El modelo que se utilizará para conocer el efecto de la calidad del gobierno y el gasto de este en el crecimiento es el modelo aumentado de Solow de Mankiw-Romer-Weil (MRW) visto en la sección anterior, pero a diferencia del ensayo original, los resultados se calcularan mediante una regresión de panel. 
La siguiente es la especificación de la ecuación: 

(15)  $ln⁡ (Y/N)=a_0+a_(1 )  ln⁡(S_K/(ngd))+a_2  ln⁡(S_H/ngd)+a_3  ln S_G/ngd+a_4  magobernanza+ a_5  agobernanza+ a_6  bgobernanza $

La descripción de las variables es		
* Y/N: Se utilizó la producción per cápita en lugar de la producción por trabajador al probar el modelo Solow. Mientras se pruebe el paradigma de Solow, se puede emplear la variable per cápita o por trabajador. MRW utilizó la producción por trabajador, mientras que Islam (1995) y Caselli (1996) utilizaron la variable per cápita.
* Sk: Porcentaje de la formación bruta de capital en el PBI real con base en la paridad de poder adquisitivo actual.
* Sh: Índice de capital humano por persona, basado en años de escolaridad y retornos a la educación. Es un proxy para la inversión en capital humano.
* Sg: Porcentaje del consumo del gobierno en el PBI real con base en la paridad de poder adquisitivo actual.
* ngd: es la suma de la tasa media de depreciación del stock de capital (d), la tasa de crecimiento de la tecnología (g) y la tasa de crecimiento de la población (n).  En este trabajo se supone que la suma de la tasa de crecimiento de la tecnología, g, y la tasa de depreciación, s, es 0.05 como en MRW. Si bien esta suposición puede no parecer apropiada para todos los países de la muestra, proporciona un punto de referencia útil para probar el modelo en cuestión. Este supuesto se ha utilizado en el trabajo de Aschauer (2000), Hulten (1996), Nazrul (1995) entre otros, tanto para las economías desarrolladas como en desarrollo.
* Variables Dummy de Calidad de Gobierno: se basan en la base de datos de gobierno compilada por Kaufmann, Kraay y Mastruzzi (2006) y que forma parte del World Bank's World Development Indicators. Kaufmann y col. han construido seis indicadores de gobernanza: (1) voz y responsabilidad: el grado en que los ciudadanos de un país pueden participar en el proceso de toma de decisiones políticas (2) estabilidad política y ausencia de violencia: mide la estabilidad de un gobierno ante la violencia política y el terrorismo (3) la efectividad del gobierno mide la capacidad de un gobierno para implementar políticas efectivas y mantener credibilidad (4) la calidad regulatoria es la capacidad del gobierno para formular e implementar políticas sólidas que fomenten la participación del sector privado (5) el imperio de la ley es la existencia de un buen sistema legal que incluye los derechos de propiedad y la ejecución de los contratos (6) el control de la corrupción mide el grado en que el poder público se desvía del beneficio privado. Las variables dummy de calidad de gobierno en el presente estudio se construyen de la siguiente manera. Los indicadores individuales (1, 2, 3, 4, 5 y 6) para cada año se promedian para construir un índice de gobernanza global compuesto (índice de calidad). Este índice de gobernanza compuesto varía de un valor de -2.5 a +2.5. Para distinguir entre los diferentes niveles de gobierno de muy alto a muy bajo, se identifican cuatro variables dicotómicas de gobierno:
	Θ ≥ 1.5 muy alta calidad de gobierno(magobernanza)
	1.5 > Θ > 0 alta calidad de gobierno (agobernanza)
	0 > Θ > -1.5 baja calidad de gobierno (bgobernanza)
	 Θ ≤ -1.5 muy baja calidad de gobierno (mbgobernanza)
	Donde Θ es el índice de calidad de gobierno compuesto.

Como la regresión de MCO no considera los efectos específicos del país no observados en el modelo (ec. 15), por lo tanto, la heterogeneidad del país puede aparecer en los parámetros estimados. En consecuencia, se calculará la ecuación 15 con efectos específicos de los países no observados usando last squeare dummy variable (LSDV) para comparar con los resultados con la regresión pooled. El uso de efectos no observados específicos de los países tiene una gran ventaja porque reduce los errores de medición.


```python
# Importamos las librerías
import pandas as pd
import numpy as np
import math as mat
import statsmodels as sm
import statsmodels.api as sma
import statsmodels.formula.api as smf
import seaborn as sns
import scipy.stats as scs
from linearmodels import PanelOLS, FirstDifferenceOLS
from linearmodels.panel import compare

# Creamos un DataFrame en Pandas con los datos de la PWT que abarca desde 2002 hasta 2017
PWT= pd.read_csv("pwt91final.csv", sep=';')
PWT.columns
```

```python
# Indexamos el Dataframe segun paises y años 
PWT = PWT.set_index(['country','year'])
PWT
```

```python
# Renombramos y creamos variables necesarias para el modelo
PWT['sk'] = PWT['csh_i']
PWT['sh'] = PWT['hc']
PWT['sg'] = PWT['csh_g']

PWT['ngd'] = PWT['n'] + 0.05

PWT['sk/ngd']= PWT['sk']/PWT['ngd']
PWT['sh/ngd']= PWT['sh']/PWT['ngd']
PWT['sg/ngd']= PWT['sg']/PWT['ngd']

# Armamos las variables Dummy de calidad de Gobierno 
PWT['mbgovernanza'] = 0 
PWT.loc[(PWT.gobindex<-1.5), 'mbgovernanza'] = 1

PWT['bgovernanza'] = 0
PWT.loc[(PWT.gobindex>=-1.5) & (PWT.gobindex<0), 'bgovernanza']= 1

PWT['agovernanza'] = 0
PWT.loc[(PWT.gobindex>=0) & (PWT.gobindex<=1.5), 'agovernanza']= 1

PWT['magovernanza'] = 0
PWT.loc[ (PWT.gobindex>1.5), 'magovernanza']= 1
```

```python
# Realizamos la regresion Pooled
pooled = PanelOLS.from_formula("np.log(PWT['Y/N'])~ 1 + np.log(PWT['sk/ngd']) + np.log(PWT['sh/ngd']) + np.log(PWT['sg/ngd']) + PWT['magovernanza'] + PWT['agovernanza'] + PWT['bgovernanza'] ", data=PWT).fit(cov_type="robust", use_lsdv=True)
print(pooled)
```

El panel de la tabla 1 muestra los resultados del modelo. En la regresión de panel pooled todas las variables son significativas con un nivel de significancia de 0,01. Las variables dummy de calidad de gobierno donde el grupo de referencia es el de muy baja calidad, muestran que mientras mayor sea la gobernanza de los países mayor será el ingreso per capita de estos. Y los coeficientes estimados sugieren que los países con “muy alta” calidad de gobierno posen un [100(e1,76-1)] ≈ 481,17% más de ingresos per cápita que los países con “muy mala” gobernanza, aquellos con “alta” calidad de gobierno disponen un 297,49% más que estos y por ultimo las economías con “baja” gobernanza cuentan con un 68,20% mayor de producción per capita que el grupo de peor calidad de gobierno. La elasticidad del ingreso per capita respecto a el consumo del gobierno es negativa y significaría que un aumento de este en un 1% disminuye PBI real per cápita en 0,81%. Los coeficientes estimados del capital privado y el capital humano son positivos e indican que un incremento del 1% del primero aumenta el ingreso per cápita 0,52% mientras que el segundo adiciona un 1,5%. 

```python
# Realizamos la regresion con efectos fijos LSDV
lsdv = PanelOLS.from_formula("np.log(PWT['Y/N'])~ 1 + EntityEffects  + np.log(PWT['sk/ngd']) + np.log(PWT['sh/ngd']) + np.log(PWT['sg/ngd']) + PWT['magovernanza'] + PWT['agovernanza'] + PWT['bgovernanza'] ", data=PWT).fit(cov_type="robust",use_lsdv=True)
print(lsdv)
```

La tabla 2 muestra los resultados usando los estimadores LSDV para incorporar los efectos no observables de los países en la regresión. Todas las variables poseen significancia para un nivel de 0,01. Al igual que en la regresión pooled las variables dummy muestran que cuanto mayor sea la calidad de gobierno mayor será el ingreso per cápita de los países ya que sus coeficientes sugieren que los países con gobernanza 'muy alta' sus ingreso per cápita son [100 (e0.3954-1)] ≈ 48,50% mayores que los países con gobernanza 'muy baja', y aquellos con gobernanza 'alta' poseen  45.60% más de producción per cápita , y los países con 'baja' ' calidad de gobierno disponen  39,78% más que los países con una gobernanza 'muy baja '. La elasticidad del ingreso per capita respecto a el consumo del gobierno es negativa y significaría que un aumento de este en un 1% disminuye PBI real per cápita en 0,07%. Los coeficientes estimados del capital privado y el capital humano son positivos e indican que un incremento del 1% del primero aumenta el ingreso per cápita 0,12% mientras que el segundo adiciona un 0,48%

Este estudio examina los efectos del tamaño y la calidad del gobierno sobre el desarrollo económico en 115 países para el periodo 2002 al 2017.
